```python
# 1. Entrada e Saída
# Lê um dado do usuário e armazena na variável 'nome'
nome = input("Digite seu nome: ")
# Exibe uma mensagem na tela
print("Olá,", nome)
```

```python
# 2. Tipos de Dados
# Número inteiro
x = 10
# Número decimal (float)
y = 3.14
# Texto (string)
texto = "Python"
# Valor booleano (True ou False)
ativo = True
```

```python
# 3. Estruturas de Controle
# Condicional: Verifica se a idade é maior ou igual a 18
idade = 20
if idade >= 18:
    print("Maior de idade")  # Se a condição for verdadeira, exibe esta mensagem
else:
    print("Menor de idade")  # Se for falsa, exibe esta mensagem
```

```python
# Laço 'for' que percorre números de 0 a 4
for i in range(5):
    print(i)  # Exibe o número atual
```

```python
# Laço 'while' que repete enquanto x for menor que 5
x = 0
while x < 5:
    print(x)  # Exibe o valor de x
    x += 1  # Incrementa x em 1 a cada iteração
```

```python
# 4. Estruturas de Dados
# Lista (array mutável)
numeros = [1, 2, 3, 4]
numeros.append(5)  # Adiciona o número 5 à lista
```

```python
# Tupla (imutável)
coordenadas = (10, 20)
```

```python
# Dicionário (estrutura chave-valor)
aluno = {"nome": "Carlos", "idade": 20}
print(aluno["nome"])  # Acessa o valor associado à chave "nome"
```

```python
# Conjunto (não aceita valores duplicados)
cores = {"vermelho", "azul", "verde"}
cores.add("amarelo")  # Adiciona um novo item ao conjunto
```

```python
# 5. Funções
# Definição de uma função chamada 'soma' que recebe dois números como argumentos
def soma(a, b):
    return a + b  # Retorna a soma dos dois números

# Exemplo de uso da função
resultado = soma(5, 3)  # Soma 5 + 3
print("O resultado da soma é:", resultado)  # Exibe o resultado: 8
```

```python
# 5.1 Função lambda (função anônima que retorna o quadrado de um número)
quadrado = lambda x: x ** 2
print(quadrado(4))  # Retorna 16
```

```python
# 6. Manipulação de Arquivos
# Abre um arquivo para leitura ('r' = read)
with open("dados.txt", "r") as arquivo:
    conteudo = arquivo.read()  # Lê o conteúdo do arquivo
    print(conteudo)  # Exibe o conteúdo lido
```

```python
# 6.1 Abre um arquivo para escrita ('w' = write) e escreve um texto nele
with open("saida.txt", "w") as arquivo:
    arquivo.write("Escrevendo no arquivo!")
```

```python
# 7. Manipulação de Exceções
try:
    x = 10 / 0  # Isso gera um erro (divisão por zero)
except ZeroDivisionError:
    print("Erro: divisão por zero!")  # Captura e trata o erro
```

```python
# 8. Módulos e Bibliotecas
# Importa o módulo math (matemática)
import math
print(math.sqrt(25))  # Retorna a raiz quadrada de 25

# Instalação de pacotes com pip (comando no terminal)
# pip install requests
```
