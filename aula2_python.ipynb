```python
import pandas as pd  # Importamos a biblioteca Pandas para manipulação de dados

# Passo 1: Criar um DataFrame
# Criamos um dicionário com dados fictícios
dados = {
    'Nome': ['Alice', 'Bruno', 'Carlos', 'Daniela'],
    'Idade': [25, 30, 22, 28],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Curitiba']
}
```

```python

# Convertendo o dicionário em um DataFrame
df = pd.DataFrame(dados)


```

```python
# Exibir o DataFrame inicial
print("DataFrame Inicial:")
print(df)
```

```python

# Passo 2: Acessar dados do DataFrame
print("\nAcessando coluna 'Nome':")
print(df['Nome'])  # Acessa a coluna Nome

```

```python
print("\nAcessando a primeira linha:")
print(df.loc[0])  # Acessa a primeira linha
```

```python

print("\nAcessando idade de Carlos:")
print(df[df['Nome'] == 'Carlos']['Idade'])  # Filtra a idade de Carlos

```

```python
# Passo 3: Adicionar uma nova linha ao DataFrame
novo_dado = {'Nome': 'Eduarda', 'Idade': 26, 'Cidade': 'Salvador'}
df = df.append(novo_dado, ignore_index=True)

```

```python
print("\nDataFrame após adicionar uma nova linha:")
print(df)

```

```python
# Passo 4: Excluir uma linha do DataFrame
df = df.drop(2)  # Remove a linha com índice 2 (Carlos)
df.reset_index(drop=True, inplace=True)  # Reseta os índices após remoção

```

```python
print("\nDataFrame após excluir Carlos:")
print(df)

```

```python
# Passo 5: Modificar um valor específico
df.at[1, 'Idade'] = 31  # Modifica a idade de Bruno para 31

```

```python
print("\nDataFrame após modificar a idade de Bruno:")
print(df)

```

```python
# Passo 6: Adicionar uma nova coluna ao DataFrame
df['Profissão'] = ['Engenheira', 'Advogado', 'Médica', 'Arquiteta']

```

```python
print("\nDataFrame após adicionar a coluna 'Profissão':")
print(df)

```

```python
# Passo 7: Ordenar o DataFrame por idade
df = df.sort_values(by='Idade')

```

```python
print("\nDataFrame ordenado por idade:")
print(df)

```

```python
# Passo 8: Salvar o DataFrame em um arquivo CSV
df.to_csv('dados_pessoas.csv', index=False)
print("\nArquivo 'dados_pessoas.csv' salvo com sucesso!")
```
