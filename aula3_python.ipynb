```python
import pandas as pd  # Biblioteca para manipulação de dados em tabelas
import numpy as np  # Biblioteca para operações matemáticas e manipulação de arrays
import matplotlib.pyplot as plt  # Biblioteca para visualização de dados e criação de gráficos
```

```python
# Criando um DataFrame de exemplo
dados = {
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eduardo'],
    'Idade': [23, 35, 30, 27, 40],
    'Salario': [2500, 4000, 3000, 3200, 5000],
    'Departamento': ['RH', 'TI', 'TI', 'RH', 'Financeiro']
}
```

```python
df = pd.DataFrame(dados)
print("DataFrame Inicial:")
print(df, "\n")
```

```python
# Estatísticas básicas
descricao = df.describe()
print("Estatísticas básicas:")
print(descricao, "\n")
```

```python
# Filtrando dados
salario_maior_que_3000 = df[df['Salario'] > 3000]
print("Funcionários com salário maior que 3000:")
print(salario_maior_que_3000, "\n")
```

```python
# Adicionando uma nova coluna
df['Bonus'] = df['Salario'] * 0.10
print("DataFrame com coluna de Bônus:")
print(df, "\n")
```

```python
# Excluindo uma linha do DataFrame
df = df.drop(index=2)
print("DataFrame após exclusão de linha:")
print(df, "\n")
```

```python
# Resetando o índice após a exclusão
df = df.reset_index(drop=True)
print("DataFrame com índice resetado:")
print(df, "\n")
```

```python
# Incluindo uma nova linha no DataFrame
nova_linha = pd.DataFrame([{'Nome': 'Fernanda', 'Idade': 29, 'Salario': 4500, 'Departamento': 'Marketing', 'Bonus': 450}])
# Usando pd.concat() em vez de append(), pois append() foi removido nas versões mais recentes do pandas
df = pd.concat([df, nova_linha], ignore_index=True)
print("DataFrame após inclusão de nova linha:")
print(df, "\n")
```

```python
# Salvando em um arquivo CSV
df.to_csv("dados_funcionarios.csv", index=False)
print("Arquivo CSV salvo!\n")
```

```python
# Carregando o CSV
novo_df = pd.read_csv("dados_funcionarios.csv")
print("DataFrame carregado do CSV:")
print(novo_df, "\n")
```

```python
# Gráfico de salários
plt.figure(figsize=(8, 5))  # Define o tamanho da figura do gráfico
plt.bar(novo_df['Nome'], novo_df['Salario'], color='blue')  # Cria um gráfico de barras com os nomes no eixo X e os salários no eixo Y
plt.xlabel('Nome')  # Adiciona rótulo ao eixo X
plt.ylabel('Salário')  # Adiciona rótulo ao eixo Y
plt.title('Salário dos Funcionários')  # Adiciona um título ao gráfico
plt.show()  # Exibe o gráfico
```

```python
# Gráfico de salários
plt.figure(figsize=(8, 5))
# Ordenando os dados pelo salário antes de plotar
df_sorted_graph = df.sort_values(by='Salario')
plt.bar(df_sorted_graph['Nome'], df_sorted_graph['Salario'], color='blue')
plt.xlabel('Nome')
plt.ylabel('Salário')
plt.title('Salário dos Funcionários')
plt.show()
```

```python
# Ordenando o DataFrame por idade
df_sorted = novo_df.sort_values(by='Idade')
print("DataFrame ordenado por idade:")
print(df_sorted, "\n")
```

```python
# Contagem de valores únicos em uma coluna
contagem_departamento = df_sorted['Departamento'].value_counts()
print("Contagem de funcionários por departamento:")
print(contagem_departamento, "\n")
```
